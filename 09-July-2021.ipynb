## Matplotlib:

- It is a plotting library.
- We can represent the data in the form of a graph or a picture by using this package.
- **pyplot** is the main module that can help us to draw graphs.


**Different Types of Plots**
- Line plot
    - Linear graph
    - Non-linear graph
- Bar Graph
- Scatter plot
- Histogram
- Pie Chart

```python
## Linear Graph
x = [5,8]
y = [4,9]
from matplotlib import pyplot as plt
# for plotting function is plot()
plt.plot(x,y)

plt.text(5,4,'(5,4)')
plt.text(8,9,'(8,9)')

plt.show()
```

```python
## EXAMPLE:1
n = list(range(1,7))
n2 = [i**2 for i in x]
#print(x)
#print(y)
plt.xlabel("X values")
plt.ylabel("Y values")

#plt.title("Graph b/w X values and it's square values")
plt.title("Non-Linear Graph")
plt.plot(n,n2,linestyle='-.',c='m')
plt.plot(n2,n)
plt.show()
```

```python
plt.plot(n, n2, 'g*-.', lw=2, ms=12,mec='red',mew=2)
#plt.grid()
```
