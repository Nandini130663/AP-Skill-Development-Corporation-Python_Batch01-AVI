```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,4,30)
y = np.logspace(0,4,30)
print(x)
print(y)
```

```python
plt.plot(x,y,c = 'g',label='linspace values')
plt.plot(x**2,y*3,color = 'm',label='logspace values')
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(loc = 'center left')
plt.show()
```

```python
plt.plot(x,y,color = 'g',label = "Line plot -1")
plt.plot(x**2,y*3,color = 'm',label = "line plot -2")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend(loc = "center")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.array([1, 2, 3, 4])
y = x*2
  
plt.plot(x, y)
  
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
  
plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
  
plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```

**Scatter plot**:
- Scatter plots are widely used to represent relation among variables and how change in one affects the other.

* scatter()

- ***x_axis_data***- An array containing x-axis data
- ***y_axis_data***- An array containing y-axis data
- ***s*** -marker size- (can be scalar or array of size equal to size of x or y)
- ***c*** -color of sequence of colors for markers
- ***marker***- marker style
- ***cmap***- cmap name
- ***linewidth***- width of marker border
- ***edgecolor***- marker border color
- ***alpha***- blending value, between 0 (transparent) and 1 (opaque)

```python
x = np.arange(1,50,5)
y = np.random.randint(1,50,size = len(x))
print(x)
print(y)
```

```python
x = [1,2,3,4,5]
y = [3,3,3,3,3]
plt.scatter(x,y,label="x vs y")
plt.scatter(y,x,label="y vs x")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
  
# dataset-1
x1 = [89, 43, 36, 36, 95, 10, 
      66, 34, 38, 20]
  
y1 = [21, 46, 3, 35, 67, 95, 
      53, 72, 58, 10]
  
# dataset2
x2 = [26, 29, 48, 64, 6, 5,
      36, 66, 72, 40]
  
y2 = [26, 34, 90, 33, 38, 
      20, 56, 2, 47, 15]
  
plt.scatter(x1, y1, c ="pink", 
            linewidths = 2, 
            marker ="s", 
            edgecolor ="green", 
            s = 50)
  
plt.scatter(x2, y2, c ="yellow",
            linewidths = 2,
            marker ="^", 
            edgecolor ="red", 
            s = 200)
  
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**Bar Plot**:

- A bar chart describes the comparisons between the discrete categories. 
- One of the axis of the plot represents the specific categories being compared,
- while the other axis represents the measured values corresponding to those categories.
    * Syntax: plt.bar(x, height, width, bottom, align)

```python
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,width = 4)
plt.show()
```

```python
plt.bar(x,y,width = 2,color = "g")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
data = {'mango':20, 'guava':15, 'apple':30,'banana':10,'grape':25,
        'kiwi':35}
fruits = list(data.keys())
price = list(data.values())

# creating the bar plot
plt.bar(fruits, price, color ='maroon',
        width = 0.4)
 
plt.xlabel("Fruits Available")
plt.ylabel("Prise of each")
plt.title("Prices of Different Fruits")
plt.show()
```

```python
help(plt.bar)
```

```python
plt.bar(x,y,width = 4,color = "c",edgecolor = "0")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
data = {'mango':20, 'guava':15, 'apple':30,'banana':10,'grape':25,
        'kiwi':35}
fruits = list(data.keys())
price = list(data.values())

# creating the horizontal bar plot
plt.barh(fruits, price, color ='maroon',height=0.4)
 
plt.xlabel("Fruits Available")
plt.ylabel("Prise of each")
plt.title("Prices of Different Fruits")
plt.show()
```

**Area Plot**:

stackplot()

```python
days = [1,2,3,4]
working = [8,11,9,10]
sleeping = [8,10,11,12]
plt.stackplot(days,working,sleeping,colors = ['g','m'],labels = ['Working',"sleeping"])
plt.legend()
plt.show()
```

**pie plot**:

pie()

```python
subjects = ['python','Java','C']
Marks = [90,50,30]
plt.pie(Marks,labels = subjects,colors = ['g','m','b'],radius = 1)
plt.show()
plt.savefig("piechart.png")
```

```python
subjects = ['python','Java','C']
Marks = [90,50,30]
plt.pie(Marks,labels = subjects,colors = ['g','m','b'],radius = 1,explode = [1,1,1])
plt.show()
```
